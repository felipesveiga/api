<h3 style='font-size:30px;font-style:italic'> Componentes da URL</h3>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Domain</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Consiste no nome do domínio em que a(s) API'(s) estão disponibilizadas.
        </li>
        <li>
            No caso, ele seria usado para obter o endereço de IP do servidor que as hospeda.
        </li>
        <li>
            Muitas empresas preferem hospedar suas API's num domínio diferente de seu site oficial, com o uso de sub-domínios, por exemplo (paypal.com$\to$api.paypal.com).
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Product [Optional]</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Consiste numa camada de abstração em que agrupamos diversos recursos, com base em características em comum.
        </li>
        <li>
            Em empresas grandes, o campo de product pode ser usado para apartar diferentes API's sob responsabilidades de times distintos.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Version [Optional]</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Consiste na versão de sua API. Dar a opção ao cliente de escolhê-la impede de expô-lo a possíveis erros de atualizações.
        </li>
        <li>
            No entanto, esse identificador não precisa estar no caminho, uma vez que ele pode ser especificado como um metadado na requisição do cliente, também.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Resource</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Finalmente, o recurso um tipo de entidade hospedado pela API com a qual conseguimos interagir. 
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> Practices for Resource Names, Actions & Associations
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Nessa aula, aprendemos algumas boas-práticas nas nomenclaturas de URL's de API's.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Resource Name</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Recomenda-se usarmos substantivos como nomes de entidades tidas pela API. Além disso, costuma-se usá-los no plural.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Resource Association</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            A rota de um determinado recurso pode ser usada para encapsular outras entidades relacionadas a ele.
        </li>
        <li>
            Nessa situação, esses sub-recursos podem ser interpretados como atributos da entidade raiz.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Action Hosting</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Podemos fazer com que a API hospede não apenas entidades, mas também ações que podem ser realizadas no backend. Neste caso, recomenda-se definir o nome do recurso como um verbo.
        </li>
        <li>
            O exemplo mais notório é o da API do Google, em que todas as buscas são feitas no recurso "search".
        </li>
        <li>
            É indicado habilitarmos rotas de ação quando uma entidade puder se consultada usando diferentes parametros. Por exemplo, numa API do LinkedIn, poderíamos buscar por pessoas por nome, localização ou empresa atual (search?name=&lt;name&gt;&location=&lt;location&gt;&employer=&lt;employer&gt;).
        </li>
    </ul>
</div>
