<h2 style='font-size:30px'> API Lifecycle & Developer Productivity
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            De maneira geral, podemos separar o ciclo de vida de uma API em quatro partes.
            <ul> 
                <li>
                    <i> Build:</i> Montagem da API, ou de uma nova versão.
                </li>
                <li>
                    <i> Publish:</i> Publicação da versão da API.
                </li>
                <li>
                    <i> Deprecated:</i> Categorização da versão como defasada, permitindo o seu uso apenas para os usuários que já tinham acesso a ela, até uma data acordada.
                </li>
                <li>
                    <i> Retired:</i> Remoção da versão dos servidores.
                </li>
            </ul>
        </li>
    </ul>
</div>
