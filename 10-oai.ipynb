<h2 style='font-size:30px'> Swagger/OAI Specifications
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Nessa aula, iniciamos o aprendizado sobre como escrever contratos da especificação Open API.
        </li>
        <li>
            Um contrato Swagger é composto por seções pré-estabelecidas, nós passaremos por cada uma delas.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> info</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            É onde especificamos os metadados da API. Possui tanto chaves simples, quanto compostas.
        </li>
        <li>
            Entre as simples, podemos mencionar:
            <ul> 
                <li>
                    <i> title</i>
                </li>
                <li>
                    <i> description</i>
                </li>
                <li>
                    <i> version</i>
                </li>
            </ul>
        </li>
        <li>
            Já as compostas seriam as abaixo:
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> contact</h4>
<div> 
    <ul style='font-size:20px'>
        <li> 
            Chave com informações de contato dos criadores da API. Essa pode ter os seguintes campos:
            <ul>
                <li>
                    <i> name:</i> Nome do criador.
                </li>
                <li>
                    <i> url:</i> URL de site do criador.
                </li>
                <li>
                    <i> email:</i> Email do criador.
                </li>
            </ul>
        </li>
     </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> license</h4>
<div> 
    <ul style='font-size:20px'>
        <li> 
            Essa chave disponibiliza informações sobre a licença da API.
            <ul>
                <li>
                    <i> name:</i> Nome da licença.
                </li>
                <li>
                    <i> url:</i> URL da licença.
                </li>
            </ul>
        </li>
     </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> host</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Domain Name da API, junto com o número da porta onde as interações com os clientes vai ocorrer.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> basePath</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Caminho-base da API, a partir do qual serão especificados todos os seus recursos.            
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> schemes</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Listagem de todos os protocolos de comunicação utilizados pela API.
        </li>
        <li>
            Note que em produção, é fortemente recomendado se evitar o uso de HTTP.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> consumes</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Especificação dos tipos de documentos admitidos como inputs pela API (.json, .xml...)
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> produces</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Especificação dos tipos de documentos admitidos como outputs pela API (.json, .xml...).
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> tags</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            O campo tags contém um array de mapeamentos chave-valor usados para oferecer mais detalhes sobre a API.
        </li>
        <li>
            As tags normalmente são referenciadas nos campos <i> paths</i> e <i> operations</i> para dar maior explicabilidade sobre as rotas da API.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic;text-decoration:underline'> externalDocs</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Campo em que podemos especificar outras documentações para melhor entendimento da operação sendo feita.
        </li>
        <li>
            Conseguimos, por exemplo, darmos um link para a documentação de FAQ dela. 
        </li>
        <li>
            Esse seria composto pelos seguintes campos:
            <ul> 
                <li>
                    <i> description:</i> Descrição sobre a documentação externa.
                </li>
                <li>
                    <i> url:</i> URL da documentação externa.
                </li>
            </ul>
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> paths</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            A seção <i> paths</i> especifica as rotas de nossa API, detalhando todas as operações HTTP suportadas, assim como os códigos de erro que podem ser apresentados.
        </li>
        <li>
            Para cada rota, podemos fornecer informações sobre o fucionamento dos métodos HTTP suportados.
        </li>
        <li>
            Recomenda-se especificá-la, antes da <i> definitions</i>.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> operação (get,post,...)</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Chave com o nome da operação. É dentro dela que vamos especificar todos os detalhes para poder executá-la na rota em questão.
        </li>
    </ul>
</div>

<h5 style='font-size:25px;font-style:italic;text-decoration:underline'> summary</h5>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Título sobre o que do que a operação se trata na rota.
        </li>
    </ul>
</div>

<h5 style='font-size:25px;font-style:italic;text-decoration:underline'> description</h5>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Texto com a descrição sobre a operação, num nível um pouco mais detalhado do que o <i> título</i>.
        </li>
    </ul>
</div>

<h5 style='font-size:25px;font-style:italic;text-decoration:underline'> tags</h5>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Lista de tags que podemos usar para caracterizar a operação.
        </li>
    </ul>
</div>

<h5 style='font-size:25px;font-style:italic;text-decoration:underline'>externalDocs</h5>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Campo em que podemos especificar outras documentações para melhor entendimento da operação.
        </li>
        <li>
            Esse seria composto pelos seguintes campos:
            <ul> 
                <li>
                    <i> description:</i> Descrição sobre a documentação externa.
                </li>
                <li>
                    <i> url:</i> URL da documentação externa.
                </li>
            </ul>
        </li>
    </ul>
</div>

<h5 style='font-size:25px;font-style:italic;text-decoration:underline'> consumes</h5>
<div> 
    <ul style='font-size:20px'> 
        <li>
              MIME types admitidos na requisição para a aplicação (application/json, text/html). 
        </li>
    </ul>
</div>

<h5 style='font-size:25px;font-style:italic;text-decoration:underline'> produces</h5>
<div> 
    <ul style='font-size:20px'> 
        <li>
              MIME types admitidos na resposta da aplicação (application/json, text/html). 
        </li>
    </ul>
</div>

<h5 style='font-size:25px;font-style:italic;text-decoration:underline'> parameters</h5>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Listagem de parâmetros que o método oferece para a requisição. Nós podemos usar parâmetros declarados na seção <i> parameters</i>.
        </li>
    </ul>
</div>

<h5 style='font-size:25px;font-style:italic;text-decoration:underline'> responses</h5>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Seção onde listamos todos os códigos de resposta possíveis.
        </li>
        <li>
            Ela teria os seguintes campos:
            <ul> 
                <li>
                    <i> description:</i> Descrição sobre o significado do código retornado, no contexto da API.
                </li>
                <li>
                    <i> schema:</i> Schema de dados da resposta. Note que podemos aproveitar os schemas pré-definidos na seção <i> definitions</i>.
                </li>
            </ul>
        </li>
        <li>
            É possível ainda definirmos uma resposta padrão para todas as chamadas com a chave <i> default</i>.
        </li>
    </ul>
</div>

```yaml
# Exemploe de seção `paths`
paths:
    /vacations/{id}:
    get:
        summary: Retrieves the information about a vacation package
        description: Retrieves the vacation package by its informed ID.
        tags:
            - vacations
            - packages
        externalDocs:
            description: Link to the /vacations ops details and flow
            url: http://mockurl.com
        consumes:
            - application/json
        produces:
            - application/json
        parameters:
            - $ref: "#/parameters/IdInPath"
        responses:
            "200":
                description: Gets the specific hotel
                schema:
                    $ref: "#/definitions/Vacation"
            "400":
                description: When a vacation package is NOT found
                schema:
                    $ref: "#/definitions/NotFoundError"
            default:
                description: Error
                schema:
                    $ref: "#/definitions/GeneralError"
```

<h4 style='font-size:30px;font-style:italic'> externalDocs</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Campo em que podemos especificar outras documentações para melhor entendimento da API.
        </li>
        <li>
            Conseguimos, por exemplo, darmos um link para a documentação de FAQ dela. 
        </li>
        <li>
            Esse seria composto pelos seguintes campos:
            <ul> 
                <li>
                    <i> description:</i> Descrição sobre a documentação externa.
                </li>
                <li>
                    <i> url:</i> URL da documentação externa.

            </ul>
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> definitions</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            O campo <i> definitions</i> nos permite criar schemas para requisições e respostas de nossa API.
        </li>
        <li>
            Essas definições funcionam de maneira similar a OOP, onde definimos um schema com um determinado nome, com certas propriedades.
        </li>
    </ul>
</div>

```yaml
properties:
    Vacations:
        # Mencionando os atributos que devem ser obrigatoriamente definidos.
        required:
            - name
            - description
            - type
            - destinations
        properties:
            name:
                type: string
            description:
                type: string
            type:
                type: string
                # Usamos `enum` para definir valores válidos para uma certa propriedade
                enum:
                    -resort
                    - cruise
            
            destinations:
                type: array
                items:
                    # Afirmando que o array de destinos será composto por entidades do schema `Location`.
                    $ref: "#/definitions/Location"
                minItems: 1
                maxItems: 12
            numOfNights:
                type: number
                minimum: 1
                maximum: 31
    Location:
        required:
            - city 
            - country
        properties:
            city:
                type: string
            country:
                type: string
```

<h3 style='font-size:30px;font-style:italic'> parameters</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Nesse campo, definimos os parâmetros que podem ser informados em qualquer requisição à nossa API.
        </li>
        <li>
            Para cada parâmetro, devemos especificar os seguintes campos:
            <ul>
                <li>
                    <i> name:</i> Nome do parâmetro.
                </li>
                <li>
                    <i> in:</i> Onde o parâmetro deve ser informado (body, query, header, path, formData).
                </li>
                <li>
                    <i> type:</i> Data type do parâmetro.
                </li>
                <li>
                    <i> required:</i> booleano de se o parâmetro é obrigatório ou não.
                </li>
            </ul>
        </li>
    </ul>
</div>

```yaml
parameters:
    IdInPath:
        name: id
        in: path
        type: string
        required: true
```

<h3 style='font-size:30px;font-style:italic'> secuityDefinitions</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Campo em que definimos que tipo de autenticação será usado na interação com a API, assim como a forma que as credenciais do usuário deverão ser declarados.
        </li>
        <li>
            Para cada definição de segurança, devemos informar os seguintes parâmetros:
            <ul> 
                <li>
                    <i> type:</i> Tipo de autenticação usada (basic, apiKey, oauth2).
                </li>
                <li>
                    <i> name:</i> Nome do campo onde a credencial deverá ser passada.
                </li>
                <li>
                    <i> in:</i> Onde a credencial deverá ser informada.
                </li>
            </ul>
        </li>
    </ul>
</div>

```yaml
securityDefinitions:
    KeySecurity:
        type: apiKey
        in: query
        name: api_key
```
