<h2 style='font-size:30px'> Evolution of REST/JSON API
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Essa aula busca contextualizar os estudantes sobre a história das API's, e como o padrão REST/JSON se tornou o predominante nas aplicações web.
        </li>
        <li>
            As API's surgiram como uma forma de simplificar a coleta de dados de serviços da Internet. Sua forma de funcionamento é baseada no conceito de Remote Procedure Call (RPC).
        </li>
        <li>
            O RPC é um protocolo de comunicação de software que permite um computador requerer uma ação de outro implantado numa outra rede, sem precisar saber os detalhes desta. Enquanto o comando está sendo executado, o ambiente requeridor fica suspenso até receber uma devolutiva.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> A Ascensão do REST</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            O REST se tornou a principal modalidade de API nos últimos anos por providenciar um conjunto de padrões e boas-práticas. Além disso, possui compatibilidade com uma ampla gama de formatos de dados, tornando-o compatível com diversos sistemas.
        </li>
        <li>
            Modalidades mais antigas, como o SOAP, ofereciam fraca padronização de funcionamento, o que as tornavam vulneráveis a falhas em caso de mudanças.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> Introduction to RESTful API
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Uma API RESTful deve seguir um conjunto de preceitos que garantirão o seu funcionamento padronizado compatível com diferentes sistemas. 
        </li>
        <li>
            Entre os mais importantes, está a capacidade de retornar dados em diversas formatações (JSON, XML). Para isso, o backend da API deverá ter um código que converterá o representational state da instância solicitada na forma desejada.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> Private, Public and Partner API
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Um fator determinante para a elaboração de uma API seria o seu nível de exposição. Quanto mais acessível o seu uso, mais rigorosos devem ser os controles de segurança e acesso.
            <ul> 
                <li>
                    <i> Public</i>: API's abertas para qualquer pessoa da internet. Devem ter mecanismos de segurança mais rígidos para impedir possíveis invasões.
                </li>
                <li>
                    <i> Private</i>: API's de consumo interno em empresas. Por serem acessíveis somente de dentro da rede interna, normalmente requerem menos mecanismos de segurança.
                </li>
                <li>
                    <i> Partner</i>: API's abertas à internet, mas que devem ser acessadas somente por empresas parceiras ou terceiros. Demandam um controle de acesso mais rigoroso.
                </li>
            </ul>
        </li>
    </ul>
</div>
