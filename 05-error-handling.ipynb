<h2 style='font-size:30px'> REST API Error Handling Practices
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Nessa aula, aprendemos algumas boas práticas envolvendo o retorno de mensagens de erro em API's.
        </li>
        <li>
            De maneira geral, costuma-se informar o <i>status-code</i> e <i>reason-phrase</i> padrão. Caso deseje, você pode acrescentar mais alguma informação num header customizado de prefixo "x-", ou body. 
        </li>
        <li>
            Por questões de manutenibilidade, recomenda-se usar uma quantidade limitada de status codes, em torno de 10. Foque nos códigos mais comuns (200, 201, 404...).
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> Application Error Handling Patterns
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Durante o desenvolvimento da API, seria interessante seguirmos algumas boas práticas para melhor entendimento de erros tanto pelos clientes, quanto desenvolvedores.
        </li>
        <li>
            Muitas vezes, o retorno do <i>status-code</i> e <i>reason-phrase</i> oferece uma contextualização muito pobre. Nesse sentido, é sempre indicado retornarmos hints, assim como links da documentação da API para auxílio.
        </li>
        <li>
            Seguem algumas boas práticas na declaração de erros em sua api:
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Text</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
              Oferecer um texto que indique sucintamente qual o problema ocorrido. 
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> API Status Code</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
              Uma boa prática interessante seria desenvolver os status codes próprios para a sua API. Durante debugs, eles auxiliariam na identificação do problema em específico que aconteceu.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Metadados</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Alguns metadados que conviriam estar presentes no payload de uma mensagem de erro seriam:
            <ul> 
                <li>
                    <i> Timestamp:</i> Data e horário de ocorrência do problema.
                </li>
                <li>
                    <i> Method:</i> Método HTTP usado na invocação.
                </li>
                <li>
                    <i> Path:</i> Caminho em que o erro ocorreu.
                </li>
            </ul>
        </li>
    </ul>
</div>
