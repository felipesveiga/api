<h2 style='font-size:30px'> Introduction to REST Architecture Constraints
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li>
            A popularização do padrão REST fez com que a maior parte das API's seguissem o protoclo HTTP com a transferência de dados por meio de documentos JSON.
        </li>
        <li>
            Isso gerou uma confusão no mercado de se considerar uma API como RESTful pelo mero uso de HTTP com JSON.
        </li>
        <li>
            Levando isso em conta, essa aula buscará definir quais são os critérios que caracterizam uma API verdadeiramente como RESTful.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> REST Design Rules</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            O padrão REST foi inventado por Roy Fielding numa dissertação acadêmica em 2000. 
        </li>
        <li>
            Nela, ele estabeleceu 6 condições que as API's RESTful devem respeitar.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Client-Server</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            O primeiro princípio seria a separação do código usado por cliente e servidor para execução das operações.
        </li>
        <li>
            Isso garante independência entre a interface de usuário e os procedimentos executados no backend, além do reúso das peças.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Stateless</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            As requisições enviadas pelo cliente devem conter todas as informações necessárias para que o servidor consiga desempenhar a ação desejada.
        </li>
        <li>
            Isso significa que o servidor não precisará armazenar dados do cliente entre interações, aliviando a sua memória.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Cache</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Os dados gerados pelas respostas devem rotulados como cacheáveis ou não.
        </li>
        <li>
            Havendo esse discernimento, a máquina do cliente poderá armazenar as informações reutilizáveis para próximas interações, sem onerar sua memória com dados inúteis.
        </li>
        <li>
            Ao mesmo tempo, a preservação de dados em cache leva ao menor estresse dos servidores com requisições redundantes (ainda mais no paradigma stateless imposto pelo REST), além da maior sensação de performance do usuário.
        </li>
    </ul>
</div>

<h5 style='font-size:25px;font-style:italic;text-decoration:underline'> Cache Headers</h5>
<div> 
    <ul style='font-size:20px'> 
        <li>
            O gerenciamento do cache na aplicação é feito com o uso do header `cache-control` nas respostas. Com ele, o servidor poderá gerenciar o comportamento do cache da máquina do cliente. 
        </li>
        <li>
            Ele consiste numa lista de diretivas separadas por vírgulas.
        </li>
        <li>
            Abaixo, vamos listar algumas das diretivas mais utilizadas no `cache-control` e seus significados.
            <ul> 
                <li>
                    <i>no-store:</i> Os dados de resposta <strong> não</strong> devem ser cacheados.
                </li>
                <li>
                    <i>private:</i> Os dados retornados devem ser atrelado a apenas o dispositivo do usuário.
                </li>
                <li>
                    <i>public:</i> Os dados retornados podem ser usados em qualquer cache.
                </li>
                <li>
                    <i>max-age:</i> Tempo máximo de validade dos dados no cache.
                </li>
                <li>
                    <i>expires:</i> Definição da data e hora em que o dado cacheado se tornará inválido.
                </li>
                <li>
                    <i>last-modified:</i> A data e hora da última alteração do recurso cacheado.
                </li>
                <li>
                    <i>e-tag:</i> ID associado com a resposta da API.
                </li>
            </ul>
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Uniform Interface</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            As interações entre cliente e servidor devem ocorrer de maneira igual, sem um funcionamento específico para um determinado tipo de aplicação.
        </li>
        <li>
            Isso torna a API compatível com uma ampla gama de aplicações.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Layered System</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Impor uma hierarquia em nossos sistemas leva à simplificação da comunicação entre seus componentes.
        </li>
        <li>
            Ao colocarmos um intermediário entre dois componentes, esse poderá abstrair todos os detalhes técnicos da parte requisitada, tornando a requisitante menos vulnerável a falhas por alterações internas da outra.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Code on Demand [Optional]</h4>

        <li>
            Essa restrição opcional se refere à capacidade de o servidor conseguir enviar códigos aos seus clientes, e não apenas dados.
        </li>
    </ul>

<div> 
    <ul style='font-size:20px'> 
        <li>
            O Richardson Maturity Model consiste num modelo que busca avaliar o nível de maturidade de uma API REST.
        </li>
        <li>
            Seu score vai de 0 a 3, sendo que quanto maior, melhor.
        </li>
    </ul>
