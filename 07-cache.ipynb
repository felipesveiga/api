<h2 style='font-size:30px'> API Caching Concepts & Design decisions
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Quando pensamos no uso de cache, esperamos em reduzir o estresse da nossa infraestrutura para requisições redundantes, ao mesmo tempo em que garantimos uma baixa latência na experiência de usuário.
        </li>
        <li>
            Dentro do framework prosposto por Roy Fielding, o armazenamento de informações em cache pode acontecer em qualquer componente da nossa infra. Cabe ao arquiteto definir onde elas serão depositadas. 
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Cuidados com Cache</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            A escolha pelo armazenamento ou não em cache de uma informação dependerá do seu grau de mutabilidade, assim como criticidade ao negócio.
        </li>
        <li>
            Outro ponto importante a se considerar seria o tempo de validade que o dado seria armazenado. Alguns fatores cruciais para essa decisão seriam o quão mutável a informação seria no decorrer do tempo, assim como preservação de capacity (às vezes, nem todas as informações poderão ser armazenadas em cache por escassez de recursos!). 
        </li>
        <li>
            Não valeria a pena usarmos cache no armazenamento de dados em constante alteração, como cotações de ativos.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Cache-Control</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            O armazenamento de qualquer informação em cache é regido pelo argumento <i> Cache-Control</i>. Nele, especificamos comandos separados por vírgulas.
        </li>
        <li>
            Usualmente, o servidor da API retornará as configurações de cache como um Header. No entanto, caso o usuário especifique as suas próprias preferências no header da requisição, as opções do servidor serão desconsideradas. 
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    public x private
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Os comandos <i>public</i> e <i>private</i> definem se o dado pode ser armazenado em cache componentes intermediários da chamada ou não, a depender de seu grau de sensibilidade.
        </li>
        <li>
            Caso seja definida como <i>private</i>, a informação apenas poderá ser depositada no cache do computador do cliente.
        </li>
        <li>
            Caso nenhum desses comandos seja especificado, a API entenderá a informação como pública, por padrão.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    no-cache
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            A diretiva <i> no-cache</i> significa que o dado em cache deve ser revalidado no servidor a cada nova requisição dos clientes.
        </li>
        <li>
            Repare, esse comando não significa proibição de cache!
        </li>
        <li>
            Uma forma de validar a alteração da representação seria por meio do Header <i> ETag</i>. Ele seria uma hash do conteúdo do recurso recebida após a primeira requisição feita pelo cliente. Nas próximas chamadas, esse valor será comparado com o do estado atual do recurso, a fim de verificar possibilidade de uso de cache. 
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    no-store
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Essa diretiva impede que o dado seja armazenado em cache.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    max-age
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Comano que define o tempo de validade em segundos da informação em cache.
        </li>
    </ul>
</div>
