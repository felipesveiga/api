<h2 style='font-size:30px'> Securing API with Basic Authentication
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Nessa aula, aprendemos sobre a Basic Authentication. Ela é a forma mais primordial de autenticação em API's.
        </li>
        <li>
            Ela consiste na passagem das credenciais de usuário no header Authorization. Elas seguem o formato de &lt;userame&gt;:&lt;password&gt; em Base64.
        </li>
        <li>
            No entanto, essa metodologia de autenticação é vulnerável a interceptações a ataques <i> man in the middle</i>, pelas credenciais não serem encriptadas. As chances desse tipo de ataque aumentam, principalmente pelo fato de o usuário ter que informar esses dados a cada requisição, em conformidade com o caráter stateless do REST.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> 
    <a href='https://developer.auth0.com/resources/labs/tools/jwt-basics#introduction'>Securing API with Tokens & JWT</a>
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Uma outra maneira de autenticarmos o nosso acesso a uma aplicação seria com os JSON Web Tokens (JWT).
        </li>
        <li>
            Esses tokens seriam passes temporários e com permissões definidas concedidos pelo sistema. De forma geral, o usuário deverá informar suas credenciais a um servidor de autenticação, que retornaria o JWT, em caso de sucesso.
        </li>
        <li>
            Assim, o usuário deverá informar esse token no header `Authetication` nas próximas interações para interagir com a aplicação.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Estrutura JWT</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Os JWT's são constituídos por três strings separadas por pontos. Denominamos essas partes como "Header", "Payload" e "Signature", respectivamente.
        </li>
        <li>
            Cada uma dessas strings seria um JSON com diferentes informações necessárias para a autenticação.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    Header
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            O Header é composto por um chave do tipo de token (no caso, sempre "JWT") e o algoritmo de encriptação sendo utilizado em "Signature".
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    Payload
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            O payload conteria informações referentes ao cliente. Ese JSON poderia conter 3 tipos de informações, que serão especificadas abaixo:
            <ul> 
                <li>
                    <i> Registered Claims</i>: São chaves definidas na própria especificação do JWT (RFC 7519).
                </li>
                <li>
                    <i> Public Claims</i>: São chaves definidas pela aplicação registradas no IANA Registry.
                </li>
                <li>
                    <i> Private Claims</i>: São claims específicas a uma aplicação negociadas entre o desenvolvedor e cliente.
                </li>
            </ul>
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    Signature
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            A Signature consiste numa hash formada pela concatenação das strings Base64 de Header e Payload, e um secret definido pelo servidor da aplicação.
        </li>
        <li>
            O algoritmo seria aquele especificado no campo "alg", do Header.
        </li>
        <li>
            Seu principal papel é certificar de que o JWT não foi alterado durante a sua transmissão.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> Securing API with API Key & Secret
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            A Autenticação via Key e Secret fornece ao usuário duas hashes que devem ser informadas a cada requisição.
        </li>
        <li>
            O cliente deverá primeiro se cadastrar na aplicação para receber essas strings.
        </li>
        <li>
            Caberá à aplicação decidir onde o usuário deverá informar esses dados em sua requisição (Header, Body, Query Parameters...).
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    Key
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            A key consiste num identificador único do cliente fazendo a chamada.
        </li>
        <li>
            Quando a requisição for feita, o backend deverá checar se a chave informada de fato está cadastrada na base de dados da aplicação. 
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    Secret
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            O secret seria uma string que comprove a legitimidade do cliente.
        </li>
        <li>
            Ela consiste numa hash formada pelo Header ou Payload da requisição, juntamente com uma string secreta compartilhada pela aplicação no momento do cadastro do cliente. 
        </li>
    </ul>
</div>
