<h2 style='font-size:30px'> API Lifecycle & Developer Productivity
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            De maneira geral, podemos separar o ciclo de vida de uma API em quatro partes.
            <ul> 
                <li>
                    <i> Build:</i> Montagem da API, ou de uma nova versão.
                </li>
                <li>
                    <i> Publish:</i> Publicação da versão da API.
                </li>
                <li>
                    <i> Deprecated:</i> Categorização da versão como defasada, permitindo o seu uso apenas para os usuários que já tinham acesso a ela, até uma data acordada.
                </li>
                <li>
                    <i> Retired:</i> Remoção da versão dos servidores.
                </li>
            </ul>
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Developer Portal</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Quando pensamos na Experiência de Usuário, é fundamental conseguirmos providenciar 3 recursos para nossos clientes:
            <ul> 
                <li>
                    <i> Documentação:</i> Tanto OAI, quanto de usuário.
                </li>
                <li>
                    <i> Serviço de Acesso:</i> Canal de solicitação de acesso para a API.
                </li>
                <li>
                    <i> Suporte:</i> Serviço de auxílio aos desenvolvedores com a API.
                </li>
            </ul>
        </li>
        <li>
            Todas essas funcionalidades podem ser disponibilizadas num <i> Developer Portal</i>. Conseguimos tanto desenvolver um portal interno, quanto comprar um vendido por empresas de TI.
        </li>
    </ul>
</div>

<h5 style='font-size:25px;font-style:italic;text-decoration:underline'> Documentação</h5>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Ao criar a documentação do Developer Portal, busque tornar o usuário mais o familiarizado possível com o funcionamento da API.
        </li>
        <li>
            Para isso, recomenda-se incluir features "Try It", em que ele possa interagir com a API por uma interface gráfica, ou providenciar códigos de exemplo sobre como uma requisição típica deveria ser escrita.
        </li>
        <li>
            Se possível, ainda podemos criar uma SDK que torne a interação com a API ainda mais abstraída!
        </li>
    </ul>
</div>

<h5 style='font-size:25px;font-style:italic;text-decoration:underline'> Serviço de Acesso</h5>
<div> 
    <ul style='font-size:20px'> 
        <li>
            A forma como o usuário deverá pedir acesso à sua API dependerá de suas características. Para aquelas de conjunto de clientes menor, talvez um fluxo manual e informal de pedido seja o suficiente.
        </li>
        <li>
            Agora, quando pensamos em API's para públicos mais amplos, é interessante incluirmos automações e algumas burocracias a fim de garantir que apenas as pessoas necessárias tenham acesso ao endpoint.
        </li>
        <li>
            É crucial se definir a pessoa responsável por gerenciar o permissionamento de acessos à API.
        </li>
    </ul>
</div>
