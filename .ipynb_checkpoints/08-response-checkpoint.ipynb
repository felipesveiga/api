<h2 style='font-size:30px'> Building support for Partial Responses
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Em operações GET, nem sempre é interessante que todos os dados sobre uma representação sejam retornados. Isso pode ocorrer, seja por questões de latência ou consumo de recursos, por exemplo.
        </li>
        <li>
            Por isso, podemos possibilitar <i> partial responses</i>, em que apenas certos atributos da entidade são retornados ao cliente.
        </li>
        <li>
            Normalmente, essa filtragem ocorre por meio de argumentos passados na query, como veremos abaixo.
        </li>
    </ul>
</div>

<figure>
    <center style='font-size:20px;margin-top:20px'> 
            <img src='img/08-graphql.png'>
        <figcaption style='font-size:15px'>
            O GraphQL é uma tecnologia desenvolvida pelo Facebook que pode ser usada criação de respostas parciais.
        </figcaption>
    </center>
</figure>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    fields
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Tendo funcionamento equivalente ao SELECT do SQL, <i> fields</i> exige que o usuário declare explicitamente quais atributos da representação serão retornados.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    only x ommit
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Algumas API's dão um controle mais flexível na seleção de atributos com os parâmetros <i>only</i> e <i>ommit</i>.
        </li>
        <li>
            Caso deseje apenas alguns dos dados, o usuário pode listá-los com <i>only</i>. Porém, caso deseje a maioria deles, salvo certas exceções, ele poderá usar <i> ommit</i>.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> Building support for Pagination
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Certas consultas podem retornar uma grande quantia de dados, podendo sobrecarregar o computador do cliente.
        </li>
        <li>
            Nesse cenário, as API's podem oferecer suporte à paginação de respostas, limitando o número de dados carregados pelo cliente de uma vez.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    Offset Pagination
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
             Na paginação por offset, o usuário define a instância inicial do retrieval e a quantidade de itens totais que devem ser resgatados.
        </li>
    </ul>
</div>

<figure>
    <center style='font-size:20px;margin-top:20px'> 
            <img src='img/08-offset.png'>
        <figcaption style='font-size:15px'>
            Nesse exemplo, especificamos que queremos resgatar 5 itens, a partir do item 6. Portanto, do sexto ao décimo.
        </figcaption>
    </center>
</figure>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    <a href='https://embedded.gusto.com/blog/api-pagination/'>Cursor Pagination</a>
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Nessa modalidade de paginação, temos uma entidade conhecida como "cursor", que seria um identificador da última instância resgatada pela sua requisição na base de dados consultada.
        </li>
        <li>
            O seu uso torna o resgate de informações mais eficiente, principalmente quando o cursor se refere a uma coluna-índice no banco de dados.
        </li>
        <li>
            Normalmente, usa-se o cursor junto com o parâmetro `limit`, que define quantos registros após ele vamos retornar.
        </li>
    </ul>
</div>
