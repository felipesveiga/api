<h1 style='font-size:40px'> 
    REST API Handling Change - Versioning Patterns
</h1>
<div> 
    <ul style='font-size:20px'> 
        <li>
            No decorrer do ciclo de vida de uma API, é comum termos que realizar modificações em seu funcionamento que podem impactar ou não seus consumidores.
        </li>
        <li>
            A intenção dessa seção é estudar melhores práticas em gerenciarmos essas alterações.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> Handling Changes to API
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Nessa aula, aprendemos alguns conceitos primários sobre alterações em API's.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Breaking x Non-Breaking Changes</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Uma breaking change consiste numa alteração que quebra o contrato atual da API. Isso implica que interações que não se conformarem com a modificação tenham erro, impactando a experiência de usuário.
        </li>
        <li>
            Por outro lado, existem mudanças que podem não ferir o contrato, principalmente quando se trata do acréscimo de uma nova funcionalidade, seja essa uma rota ou parâmetro opcional de consultas.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Major x Minor Changes</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            A quebra ou não do contrato da API implica na categorização de uma mudança como major ou minor.
        </li>
        <li>
            Toda modificação que cause quebra deve ser qualificada como major, e comunicada previamente para que os desenvolvedores adaptem as aplicações consumidoras. Já as minors devem se referir a alterações que não impacte a experiência de consumo.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> Versioning the API
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            O versionamento de API's deve seguir algumas boas-práticas, a fim de garantir uma experiência com o mínimo de impacto para os clientes.
        </li>
        <li>
            Antes de tudo, ao subirmos uma nova versão, devemos oferecer um período para clientes já existentes de adaptarem as suas aplicações para a consumi-la, antes de desligarmos a antiga. Com relação a novos consumidores, seria interessante orientá-los a usar a última versão disponível.
        </li>
        <li>
            Entre as formas mais comuns de se definir a versão desejada na requisição, pode-se listar:
            <ul> 
                <li>
                    HTTP Header (x-&lt;api-version-key&gt;:&lt;api-version&gt;)
                </li>
                <li>
                    Query Parameters (/posts?version=&lt;api-version&gt;)
                </li>
                <li>
                    URL path parameter (/v&lt;api-version&gt;)
                </li>
            </ul>
        </li>
    </ul>
</div>
