<h2 style='font-size:30px'> Introduction to RESTful API
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Uma API RESTful deve seguir um conjunto de preceitos que garantirão o seu funcionamento padronizado compatível com diferentes sistemas. 
        </li>
        <li>
            Entre os mais importantes, está a capacidade de retornar dados em diversas formatações (JSON, XML). Para isso, o backend da API deverá ter um código que converterá o representational state da instância solicitada na forma desejada.
        </li>
    </ul>
</div>

    <center style='font-size:20px;margin-top:20px'>

<div> 
    <ul style='font-size:20px'> 
        <li> 
            A maior parte das API's REST usam o <strong>HTTP</strong> como protocolo de comunicação, apesar desse não ser obrigatório em sua construção.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> Private, Public and Partner API
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Um fator determinante para a elaboração de uma API seria o seu nível de exposição. Quanto mais acessível o seu uso, mais rigorosos devem ser os controles de segurança e acesso.
            <ul> 
                <li>
                    <i> Public</i>: API's abertas para qualquer pessoa da internet. Devem ter mecanismos de segurança mais rígidos para impedir possíveis invasões.
                </li>
                <li>
                    <i> Private</i>: API's de consumo interno em empresas. Por serem acessíveis somente de dentro da rede interna, normalmente requerem menos mecanismos de segurança.
                </li>
                <li>
                    <i> Partner</i>: API's abertas à internet, mas que devem ser acessadas somente por empresas parceiras ou terceiros. Demandam um controle de acesso mais rigoroso.
                </li>
            </ul>
        </li>
    </ul>
</div>
