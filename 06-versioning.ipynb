<h3 style='font-size:30px;font-style:italic'> Breaking x Non-Breaking Changes</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Uma breaking change consiste numa alteração que quebra o contrato atual da API. Isso implica que interações que não se conformarem com a modificação tenham erro, impactando a experiência de usuário.
        </li>
        <li>
            Por outro lado, existem mudanças que podem não ferir o contrato, principalmente quando se trata do acréscimo de uma nova funcionalidade, seja essa uma rota ou parâmetro opcional de consultas.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Major x Minor Changes</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            A quebra ou não do contrato da API implica na categorização de uma mudança como major ou minor.
        </li>
        <li>
            Toda modificação que cause quebra deve ser qualificada como major, e comunicada previamente para que os desenvolvedores adaptem as aplicações consumidoras. Já as minors devem se referir a alterações que não impacte a experiência de consumo.
        </li>
    </ul>
</div>
