<h2 style='font-size:30px'> Building support for Partial Responses
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Em operações GET, nem sempre é interessante que todos os dados sobre uma representação sejam retornados. Isso pode ocorrer, seja por questões de latência ou consumo de recursos, por exemplo.
        </li>
        <li>
            Por isso, podemos possibilitar <i> partial responses</i>, em que apenas certos atributos da entidade são retornados ao cliente.
        </li>
        <li>
            Normalmente, essa filtragem ocorre por meio de argumentos passados na query, como veremos abaixo.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    fields
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Tendo funcionamento equivalente ao SELECT do SQL, <i> fields</i> exige que o usuário declare explicitamente quais atributos da representação serão retornados.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    only x ommit
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Algumas API's dão um controle mais flexível na seleção de atributos com os parâmetros <i>only</i> e <i>ommit</i>.
        </li>
        <li>
            Caso deseje apenas alguns dos dados, o usuário pode listá-los com <i>only</i>. Porém, caso deseje a maioria deles, salvo certas exceções, ele poderá usar <i> ommit</i>.
        </li>
    </ul>
</div>
